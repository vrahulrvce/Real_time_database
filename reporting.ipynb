```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("rintro-chapter8.csv")
print(data.head())
print(data.info()) 
print(data['brand'].value_counts())
print(data.shape)
```

```python
sample = data.groupby('brand').apply(lambda x: x.sample(frac=0.75, random_state=42)).reset_index(drop=True)
sample
```

```python
features = sample.drop('brand', axis=1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
pca_data
```

```python
explained_var = pca.explained_variance_ratio_
print("Explained Variance per Component:", explained_var)
```

```python
plt.plot(range(1, len(explained_var)+1), explained_var, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.title('Scree Plot')
plt.show()
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i}' for i in range(1, len(features.columns)+1)], index=features.columns)
print(loadings)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=sample['brand'])
for i, feature in enumerate(features.columns):
    plt.arrow(0, 0, loadings.iloc[i,0]*5, loadings.iloc[i,1]*5, color='r', alpha=0.5)
    plt.text(loadings.iloc[i,0]*5.5, loadings.iloc[i,1]*5.5, feature, color='r')
plt.xlabel(f'PC1 ({explained_var[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({explained_var[1]*100:.1f}%)')
plt.title('PCA Biplot')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
from textblob import TextBlob
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import re

# Download required NLTK data
nltk.download('vader_lexicon')
nltk.download('stopwords')

# Input text (truncated for brevity)
  # (your full text here)

# --- Sentiment Analysis ---
blob = TextBlob(text)
sentiment = blob.sentiment

sia = SentimentIntensityAnalyzer()
vader_scores = sia.polarity_scores(text)

# --- Emotion Analysis ---
emotional_words = {
    'anger': ['abuse', 'block', 'refused', 'dubious', 'overcome', 'bypass'],
    'fear': ['crisis', 'emergency', 'unforeseen', 'torture', 'attack', 'illegal'],
    'trust': ['approve', 'legal', 'intent', 'flexibility', 'constitutionality'],
    'disgust': ['partisan', 'impasse', 'override', 'veto', 'abuse']
}

emotion_counts = {emotion: sum(word.lower() in text.lower() for word in words) 
                 for emotion, words in emotional_words.items()}

# --- Word Cloud ---
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()
    stop_words = set(stopwords.words('english'))
    words = text.split()
    return ' '.join([word for word in words if word not in stop_words and len(word) > 3])

cleaned_text = clean_text(text)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(cleaned_text)

# --- Visualization ---
plt.figure(figsize=(15, 5))

# Sentiment Plot
plt.subplot(1, 3, 1)
plt.bar(['Polarity', 'Subjectivity'], [sentiment.polarity, sentiment.subjectivity], color=['blue', 'orange'])
plt.title('TextBlob Sentiment')

# Emotion Plot
plt.subplot(1, 3, 2)
plt.bar(emotion_counts.keys(), emotion_counts.values(), color='red')
plt.title('Emotion Word Frequency')
plt.xticks(rotation=45)

# Word Cloud
plt.subplot(1, 3, 3)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Key Topics Word Cloud')

plt.tight_layout()
plt.show()

# Print results
print(f"TextBlob: Polarity={sentiment.polarity:.2f}, Subjectivity={sentiment.subjectivity:.2f}")
print(f"VADER Scores: {vader_scores}")
print("Emotion Counts:", emotion_counts)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("rintro-chapter5.csv")

# Check for issues
print(df.isnull().sum())  # No missing values
print(df.duplicated().sum()) 
```

```python
df_encoded = pd.get_dummies(df, columns=['gender', 'ownHome', 'subscribe'])
df_encoded
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded[['age', 'income', 'kids']])
df_scaled = pd.DataFrame(df_scaled, columns=['age', 'income', 'kids'])
```

```python
final_df = pd.concat([df_scaled, df_encoded.drop(['age', 'income', 'kids', 'Segment'], axis=1)], axis=1)
```

```python
final_df = pd.concat([df_scaled, df_encoded.drop(['age', 'income', 'kids', 'Segment'], axis=1)], axis=1)
```

```python
sample = df.sample(frac=0.8, random_state=42)
```

```python
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(final_df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(final_df)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(final_df, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Prepare the data (using previously scaled data: final_df)
X = final_df.values

# Step 1: Determine optimal epsilon (ε) using k-distance graph
neigh = NearestNeighbors(n_neighbors=4)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)

# Plot k-distance graph to find 'elbow' point
plt.figure(figsize=(8,5))
plt.plot(np.sort(distances[:,3])[::-1])  # Using 4th nearest neighbor
plt.xlabel('Points')
plt.ylabel('4th Nearest Neighbor Distance')
plt.title('K-Distance Graph for Epsilon Estimation')
plt.grid()
plt.show()

# Based on the plot, select epsilon where the curve bends
eps = 0.5  # Adjust based on your k-distance graph

# Step 2: DBSCAN clustering
dbscan = DBSCAN(eps=eps, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(X)

# Count number of clusters (excluding noise points labeled as -1)
n_clusters = len(set(df['DBSCAN_Cluster'])) - (1 if -1 in df['DBSCAN_Cluster'] else 0)
print(f"Number of clusters found: {n_clusters}")
print(f"Number of noise points: {sum(df['DBSCAN_Cluster'] == -1)}")

# Step 3: Silhouette Score (only on non-noise points if possible)
if n_clusters > 1:
    mask = df['DBSCAN_Cluster'] != -1
    if sum(mask) > 0:  # Only calculate if we have non-noise points
        score = silhouette_score(X[mask], df.loc[mask, 'DBSCAN_Cluster'])
        print(f"Silhouette Score: {score:.3f}")
    else:
        print("Cannot compute silhouette score - all points are noise")
else:
    print("Need at least 2 clusters to compute silhouette score")

# Step 4: Visualize DBSCAN results
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='income', hue='DBSCAN_Cluster', 
                data=df, palette='viridis', legend='full')
plt.title('DBSCAN Clustering Results (Age vs Income)')
plt.xlabel('Age (Standardized)')
plt.ylabel('Income (Standardized)')
plt.show()

# Compare with original K-Means clusters
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='income', hue='Cluster', 
                data=df, palette='viridis', legend='full')
plt.title('K-Means Clustering Results (Age vs Income)')
plt.xlabel('Age (Standardized)')
plt.ylabel('Income (Standardized)')
plt.show()
```
